```python
from sklearn.datasets import make_classification
```

```python
X, y = make_classification(n_samples=10000, n_features=10, n_classes=2, weights=[0.9, 0.1], n_clusters_per_class=1, random_state=42)
```

```python
import pandas as pd
import seaborn as sns
sns.scatterplot(x=pd.DataFrame(X)[0], y=pd.DataFrame(X)[1], hue=y)
```
